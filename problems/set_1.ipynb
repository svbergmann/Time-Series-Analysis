```python


import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
from numpy import random as rand

rand.seed(1401)
```

```python
n = 120
rng = 5
```

```python
series_list = [pd.Series(rand.chisquare(size=n, df=2)) for _ in range(rng)]
```

```python
for series_enum in enumerate(series_list):
    plt.figure(series_enum[0])
    series_enum[1].plot(xlabel='Time', ylabel=f'Y_{series_enum[0]}', title=f'Time Series Plot of Y_{series_enum[0]}')
```

```python
for series_enum in enumerate(series_list):
    fig = plt.figure(series_enum[0])
    ax = plt.subplot()
    ax.set_title(f'Normal probability plot of Y_{series_enum[0]}')
    ax.set_xlabel('Theoretical Quantiles')
    ax.set_ylabel('Sample Quantiles')
    stats.probplot(series_enum[1], dist="norm", plot=plt)
```

```python
for series_enum in enumerate(series_list):
    fig = plt.figure(series_enum[0])
    ax = plt.subplot()
    ax.set_title(f'Histogram of Y_{series_enum[0]}')
    ax.set_xlabel(f'Y_{series_enum[0]}')
    ax.set_ylabel('Frequency')
    series_enum[1].hist(bins=12)
```
